# DATA AUGMENTATION

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv("insurance.csv")
```

```python
df
```

```python
df.shape
```

```python
df[['sex','smoker']] = df[['sex','smoker']].apply(LabelEncoder().fit_transform)
dum_reg=pd.get_dummies(df.region, prefix='location')
df=pd.concat([df,dum_reg],axis=1)
df=df.drop(columns="region",axis=1)
```

```python
df
```

```python
age=df['age']
sex=df['sex']
bmi=df['bmi']
children=df['children']
smoker=df['smoker']
charges=df['charges']
location_northeast=df['location_northeast']
location_northwest=df['location_southwest']
location_southeast=df['location_southeast']
location_southwest=df['location_southwest']
ages=np.std(age)
sexs=np.std(sex)
bmis=np.std(bmi)
childrens=np.std(children)
smokers=np.std(smoker)
chargess=np.std(charges)
location_northeasts=np.std(location_northeast)
location_northwests=np.std(location_northwest)
location_southeasts=np.std(location_southeast)
location_southwests=np.std(location_southwest)

df1=[]
for _, row in df.iterrows():
    temp={
        'age':row['age'],
        'sex':row['sex'],
        'bmi':row['bmi'],
        'children':row['children'],
        'smoker':row['smoker'],
        'charges':row['charges'],
        'location_northeast':row['location_northeast'],
        'location_northwest':row['location_northwest'],
        'location_southeast':row['location_southeast'],
        'location_southwest':row['location_southwest']
            
        }
    df1.append(temp)
    




for _ in range(20):
    for _, row in df.iterrows():
        temp={
            'age':row['age']+np.random.uniform(ages),
            'sex':row['sex']+np.random.uniform(sexs),
            'bmi':row['bmi']+np.random.uniform(bmis),
            'children':row['children']+np.random.uniform(childrens),
            'smoker':row['smoker']+np.random.uniform(smokers),
            'charges':row['charges']+np.random.uniform(chargess),
            'location_northeast':row['location_northeast']+np.random.uniform(location_northeasts),
            'location_northwest':row['location_northwest']+np.random.uniform(location_northwests),
            'location_southeast':row['location_southeast']+np.random.uniform(location_southeasts),
            'location_southwest':row['location_southwest']+np.random.uniform(location_southwests)
            
        }
        df1.append(temp)
      #  print(temp)
```

```python
df=pd.DataFrame(df1)
```
